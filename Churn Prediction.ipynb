### Dealing with null values

```python
#checking null values 
df.isna().sum()
```

```python
#visualizing missing values
missing=pd.DataFrame((df.isnull().sum())*100/df.shape[0]).reset_index()
plt.figure(figsize=(16,5))
ax=sns.pointplot('index',0,data=missing)
plt.xticks(rotation=90,fontsize=10)
plt.title('Percentage of missing values')
plt.ylabel('Percentage')
plt.show()
```

Observation : Here we can see some null values present in day.charge and eve.mins features.There are more than 1% null values in day.charge column and around 5% null values in evening.minutes column,So 0.62% percent total null values present in the data.

```python
#displaying where exactly null values are present.
df[df.isnull().any(axis=1)]
```

```python
#filling null values Using mean 
df['day.charge'] = df['day.charge'].fillna(df['day.charge'].median())
df['eve.mins'] = df['eve.mins'].fillna(df['eve.mins'].median())
```

#### Note: Median imputation will work better because it is a number that is already present in the data set and is less susceptible to outlier  as compared to mean imputation.

```python
# Checking null values after null value imputation
df.isnull().sum()
```

#### Above result shows that we have successfully imputed null values.

### Performance Analysis after Resampling

```python
#importing necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
import xgboost as xgb
import lightgbm as ltb
from sklearn.naive_bayes import GaussianNB as GB
from sklearn.metrics import f1_score, precision_score, recall_score, classification_report,confusion_matrix,accuracy_score, roc_auc_score, roc_curve, precision_recall_curve
from sklearn.model_selection import GridSearchCV, KFold, StratifiedKFold, cross_val_score, RandomizedSearchCV
```

#### 1.On complete data

```python
models = {
    "Logistic Regression" : LogisticRegression(),
    "Decision Tree" : DecisionTreeClassifier(),
    "Random Forest" : RandomForestClassifier(),
    "KNN" : KNeighborsClassifier(),
    "svm" : SVC(),
    "Adaboost  " : AdaBoostClassifier(),
    "XGBoost" : xgb.XGBClassifier(),
    "LGBM " : ltb.LGBMClassifier(),
    "Naive Bayes" :GB()
}
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train_sm,y_train_sm)  #train model
    
    #make prediction
    y_train_sm_pred = model.predict(x_train_sm) #on train data
    y_test_sm_pred = model.predict(x_test_norm) # on test data
    
    #training set performance 
    model_train_sm_cr = classification_report(y_train_sm,y_train_sm_pred)#classification report
    model_train_sm_cm = confusion_matrix(y_train_sm,y_train_sm_pred)#confusion matrix
    model_train_sm_ac = accuracy_score(y_train_sm,y_train_sm_pred)#accuracy score
    model_train_sm_ps = precision_score(y_train_sm,y_train_sm_pred)#precision score
    model_train_sm_rs = recall_score(y_train_sm,y_train_sm_pred)#recall score
    model_train_sm_fs = f1_score(y_train_sm,y_train_sm_pred)#f1 score
    model_train_sm_roc = roc_auc_score(y_train_sm,y_train_sm_pred)#roc_auc_score
    
    #testing  set performance 
    model_test_sm_cr = classification_report(y_test,y_test_sm_pred)#classification report
    model_test_sm_cm = confusion_matrix(y_test,y_test_sm_pred)#confusion matrix
    model_test_sm_ac = accuracy_score(y_test,y_test_sm_pred)#accuracy score
    model_test_sm_ps = precision_score(y_test,y_test_sm_pred)#precision score
    model_test_sm_rs = recall_score(y_test,y_test_sm_pred)#recall score
    model_test_sm_fs = f1_score(y_test,y_test_sm_pred)#f1 score
    model_test_sm_roc = roc_auc_score(y_test,y_test_sm_pred)#roc_auc_score
    
    
    print(list(models.keys())[i])
    
    print("Model Performance for Training set")
    print("-accuracy:{:.4f}".format(model_train_sm_ac))
    print("-precision:{:.4f}".format(model_train_sm_ps))
    print("-recall:{:.4f}".format(model_train_sm_rs))
    print("-f1 score:{:.4f}".format(model_train_sm_fs))
    print("-ROC_score:{:.4f}".format(model_train_sm_roc))
    
    print("="*30)
    
    print("Model Performance for Testing set")
    print("-accuracy:{:.4f}".format(model_test_sm_ac))
    print("-precision:{:.4f}".format(model_test_sm_ps))
    print("-recall:{:.4f}".format(model_test_sm_rs))
    print("-f1 score:{:.4f}".format(model_test_sm_fs))
    print("-ROC_score:{:.4f}".format(model_test_sm_roc))
    
    print("="*30)
    print("\n")
```
